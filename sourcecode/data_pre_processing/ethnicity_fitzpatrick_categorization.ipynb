### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Train Data

#### White

Filter out white ethnicity

```python
filter_white=['East Asian','Latino_Hispanic','Black','Indian','Southeast Asian', 'Middle Eastern']
train_white = train_data[~train_data.race.isin(filter_white)]
```

```python
train_white.head()
```

Rename white ethnicity to "I & II" phototype

```python
train_white["phototype"] = "I & II"
```

Sort 'file' in Ascending Order

```python
train_white_sorted=train_white.sort_values("file")
```

Export to CSV

(The same process is repeated for each ehtnicity in the training set and the testing set.)

```python
train_white_sorted.to_csv('/content/drive/MyDrive/fairface_resampling/fitzpatrick_categorization/train_subcategories/train_white.csv')
```
