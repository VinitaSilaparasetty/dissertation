# Fair Face Undersampling

#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load Data

```python
test_data1=pd.read_csv("/content/drive/MyDrive/fairface_resampling/test_keep_age_final.csv")
```

```python
test_data=test_data1.copy()
```

```python
test_data.head()
```

```python
test_data.shape
```

#### Undersampling

The objective is to balance the data on 'race' and 'age'. Manual undersampling was done by removing samples from the majority classes so that they are exactly equal to the minority class.

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = test_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
test_data['age'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x ='gender', data = test_data)
```

```python
test_data['gender'].value_counts()
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = test_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
test_data['race'].value_counts()
```

Taking an equal number of samples for ethnicity

```python
equal_ethnicity =test_data.groupby('race').head(979)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_ethnicity)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_ethnicity)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_ethnicity['age'].value_counts()
```

From equal ethnicities, take equal age groups

```python
equal_age_groups =equal_ethnicity.groupby('age').head(264)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_age_groups)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```
