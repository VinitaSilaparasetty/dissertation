# Fair Face Undersampling Train

#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load Data

```python
train_data1=pd.read_csv("/content/drive/MyDrive/fairfaceforprototype/train_keep_age_final.csv")
```

```python
train_data=train_data1.copy()
```

```python
train_data.head()
```

```python
train_data.shape
```

#### Undersampling data

The objective is to balance the data on 'race' and 'age'. Manual undersampling was done by removing samples from the majority classes so that they are exactly equal to the minority class.

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = train_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
train_data['age'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x ='gender', data = train_data)
```

```python
train_data['gender'].value_counts()
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = train_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
train_data['race'].value_counts()
```

Taking an equal number of samples for ethnicity

```python
equal_ethnicity =train_data.groupby('race').head(7510)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_ethnicity)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_ethnicity)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_ethnicity['age'].value_counts()
```

From equal ethnicities, take equal age groups

```python
equal_age_groups =equal_ethnicity.groupby('age').head(2212)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_age_groups)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_age_groups)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_age_groups['race'].value_counts()
```

From equal age groups take equal ethnicities

```python
equal_ethnic_2 =equal_age_groups.groupby('race').head(1229)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_ethnic_2)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_ethnic_2)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_ethnic_2['age'].value_counts()
```

From second round of equal ethnicities, take equal ages

```python
equal_age_2 =equal_ethnic_2.groupby('age').head(953)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_age_2)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_age_2)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_age_2['race'].value_counts()
```

From second round of equal ages take equal ethnicities

```python
equal_ethnic_3 =equal_age_2.groupby('race').head(599)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_ethnic_3)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_ethnic_3)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_ethnic_3['age'].value_counts()
```

```python
equal_ethnic_3.shape
```

from equal ethnicites take equal ages

```python
equal_age_3 =equal_ethnic_3.groupby('age').head(664)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='age', data = equal_age_3)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='race', data = equal_age_3)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_age_3['race'].value_counts()
```

```python
equal_age_3.shape
```

```python
equal_ethnic_3.head()
```

We take equal ethnicities as our final resampled selection. This way we ensure equal representation while minimizing the loss of data.
