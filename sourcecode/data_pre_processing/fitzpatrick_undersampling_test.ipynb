#### Undersampling

The objective is to balance the data on 'gender' and 'phototype'. Manual undersampling was done by removing samples from the majority classes so that they are exactly equal to the minority class.

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='phototype', data = test_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
test_data['phototype'].value_counts()
```

Taking an equal number of samples for phototype

```python
equal_phototype =test_data.groupby('phototype').head(1005)
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='phototype', data = equal_phototype)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='gender', data = equal_phototype)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
equal_phototype['gender'].value_counts()
```
