```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Data

```python
train_data1=pd.read_csv("/content/drive/MyDrive/fairface_resampling/fitzpatrick_categorization/fitzpatrick_type_train.csv")
```

```python
train_data=train_data1.copy()
```

```python
test_data1=pd.read_csv("/content/drive/MyDrive/fairface_resampling/fitzpatrick_categorization/fitzpatrick_type_test .csv")
```

Visualize Phototypes

```python
plt.figure(figsize=(10,10))
chart=sns.countplot(x ='phototype', data = train_data)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```
